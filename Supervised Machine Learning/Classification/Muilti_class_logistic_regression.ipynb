```python
%matplotlib inline
import matplotlib.pyplot as plt

# Load in preloaded handwritten datatset
from sklearn.datasets import load_digits
```

```python
digits =load_digits()
```

```python
# lets chack what's inside digits
dir(digits)
```

```python
# Lets look at the first image
digits.data[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
# Lets check the target
# We will use data and target to train our model
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# lets predict a random number
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
model.predict(digits.data[0:5])
```

```python
#Lets get the overall accuracy of the model
from sklearn.metrics import confusion_matrix
```

```python
# letd get the predicted values for X_test
y_predicted = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sns
plt.figure(figsize= (10,7))
sns.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
